```python
from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
df['payment_method']= label_encoder.fit_transform(df['payment_method'])
df['payment_method'].unique()
```

```python
plt.figure(figsize=(25,30))
ax = sns.heatmap(df.corr(numeric_only=True), annot=True)
```

```python
x = df[["number_of_complaints","article_category_12",]]
#target selection
y=df["churn"]
#review featureset
#X[0:5]
print(x)
```

```python

from sklearn.model_selection import train_test_split
X_trainset, X_testset, y_trainset, y_testset = train_test_split(x, y, test_size=0.25, random_state=3)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=100)
rf_model.fit(X_trainset, y_trainset)
rf_pred = rf_model.predict(X_testset)
rf_pred[0:5]
```

```python
from sklearn import metrics
rf_model.score(X_testset, y_testset)
metrics.accuracy_score(y_testset, rf_pred)
```

```python
# confusion matrics
from sklearn.metrics import confusion_matrix
confusion_matrix(y_testset, rf_pred)
cm = confusion_matrix(y_testset, rf_pred)
plt.figure(figsize=(12,6))
plt.title("Confusion Matrix")
sns.heatmap(cm, annot = True, fmt='d', cmap="rainbow")
plt.ylabel("Actual Values")
plt.xlabel("Predicted Values")
```
